```python
import os
import numpy as np
import torch
from PIL import Image
import cv2
import matplotlib.pyplot as plt
```

```python
def plot_3dbox(img, u, v, color):
    '''
    u,v: [N_box,8]
    '''
    lines = [(0,2),[1,3],(5,7),(4,6),(2,7),(0,5),(1,4),(3,6),(0,1),(2,3),(5,4),(7,6)]
    
    cv_img = img.copy()
    
    for box_u,box_v in zip(u,v):
        box_u = box_u.numpy()
        box_v = box_v.numpy()
        for line in lines:
            (pt1 ,pt2) = line
            cv2.line(cv_img, (box_u[pt1], box_v[pt1]), (box_u[pt2], box_v[pt2]), color,1)

    return cv_img
```

```python
def cam2image(cam_cood, K):
    '''
    K: [3, 3]
    cam_cood: [N_box, 8, 3]
    '''
    K = K.to(cam_cood)
    projected_pts = torch.matmul(K[None,None], cam_cood[..., None]).squeeze(-1)
    
    z = projected_pts[:,:, 2]
    z[z==0] = 1e-7
    u = torch.div(projected_pts[:,:,0], z, rounding_mode='trunc').long()
    v = torch.div(projected_pts[:,:,1], z, rounding_mode='trunc').long()

    return u,v,z
```

```python
data_root = '/data/usr/gbc/workspace/nerf_explore/PanopticNeRF/data/result/panopticnerf/panopticnerf_test_max20/2013_05_28_drive_0000_sync'
frame_id = 3385
image_file = os.path.join(data_root,'img{:04d}_gt.png'.format(frame_id))
im = Image.open(image_file)
im = np.array(im)
pred_path = os.path.join(data_root, 'img{:04d}_pred_3dbox.pt'.format(frame_id))
print(pred_path)
pred_resultss = torch.load(pred_path)

pred_3dbox_cam = pred_resultss['camera']
pred_3dbox_world = pred_resultss['world']
gt = pred_resultss['label']
K = pred_resultss['K']
pose = pred_resultss['pose']
```

```python
T = pose[:3,3].to(gt)
R = pose[:3, :3].to(gt)
R_inv = torch.inverse(R)[None,None]
gt_cam = torch.matmul(R_inv, (gt-T).unsqueeze(-1)).squeeze(-1)

u_gt, v_gt, z_gt = cam2image(gt_cam, K)
u_pred, v_pred, z_pred = cam2image(pred_3dbox_cam, K)

class_colors = {
    0:(0,255,0),
    1:(255,255,0),
    2:(0,255,255)
}
color = class_colors[0]

gt_img = plot_3dbox(im, u_gt, v_gt, color)
pred_img = plot_3dbox(im, u_pred, v_pred, color)
```
